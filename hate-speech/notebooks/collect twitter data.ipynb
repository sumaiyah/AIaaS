# Founta et al Dataset

```python
founta_labels = pd.read_csv('founta_labels.csv')

tweets = []
labels = []

# n = 1291 + 497
# founta_labels = founta_labels[n:n+500]
for i in range(len(founta_labels.values)):
    tweet_id = founta_labels.values[i][0]
    label = founta_labels.values[i][1]
    
    try:
        # authorization of consumer key and consumer secret
        auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
        # set access to user's access key and access secret 
        auth.set_access_token(access_token, access_token_secret)
        # calling the api 
        api = tweepy.API(auth)
#         api = tweepy.API(auth,wait_on_rate_limit=True,wait_on_rate_limit_notify=True)
        # fetching the status
        status = api.get_status(tweet_id, tweet_mode='extended')
        
        tweets.append(status.full_text)
        labels.append(label)
        
        print(i, end=', ')  
    except tweepy.TweepError as e:
        # ignore tweets that cant be found
        pass
```

```python
# founta_df = pd.DataFrame({'tweet': [], 'label': []})
# founta_df.to_csv('founta.csv', index=False)

founta_df = pd.read_csv('founta.csv')
update_df = pd.DataFrame({'tweet': tweets, 'label': labels})
founta_df = pd.concat([founta_df, update_df])
founta_df.to_csv('founta.csv', index=False)
```

```python
tweets = []
labels = []
i = 0

with open('waseem_labels.txt','r') as file:
    rows = file.readlines()[3169:]
    for row in rows:
        tweet_id = row.split(',')[0]
        label   = row.split(',')[1]

        try:
            # authorization of consumer key and consumer secret
            auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
            # set access to user's access key and access secret 
            auth.set_access_token(access_token, access_token_secret)
            # calling the api 
            api = tweepy.API(auth)
            #api = tweepy.API(auth, wait_on_rate_limit=True,wait_on_rate_limit_notify=True)
            # fetching the status
            status = api.get_status(tweet_id, tweet_mode='extended')

            tweets.append(status.full_text)
            labels.append(label)

            print(i, end=', ')  
            i += 1
        except tweepy.TweepError as e:
            # ignore tweets that cant be found
            pass
```
