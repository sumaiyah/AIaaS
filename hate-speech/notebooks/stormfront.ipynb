# Stormfront dataset

```python
from os import listdir
import pandas as pd
```

```python
annotations = pd.read_csv('hate-speech-dataset/annotations_metadata.csv')
print(annotations['label'].value_counts())
annotations
```

```python
annotations = annotations[annotations.label.isin(['hate', 'noHate'])]
annotations
```

```python
posts = []
labels = []

for file_id, label in annotations[['file_id', 'label']].values:
    file_path = 'hate-speech-dataset/all_files/' + '%s.txt' % file_id
    with open(file_path, 'r', encoding="utf8") as file:
        print(file_path)
        file_contents = file.read()
        
        posts.append(file_contents)
        labels.append(label)
```

```python
pd.DataFrame({'post': posts, 'label':labels}).to_csv('stormfront_data.csv', index=False)
```
